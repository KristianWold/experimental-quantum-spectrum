## Annealing

```python
n = 2
d = 2**n
T = 1

tf.random.set_seed(42)
np.random.seed(42)

H_target = SpinSpin()
H_model = SpinSpin()

channel_target = MagnusPropagator(liouvillian=H_target, grid_size=100)
channel_model1 = MagnusPropagator(liouvillian=H_model, grid_size=100)
print(state_fidelity(channel_target.choi(T), channel_model1.choi(T))/16)
```

```python
T = 1
optimizer = tf.optimizers.Adam(learning_rate=0.05)
loss1 = []
for i in tqdm(range(1000)):
    with tf.GradientTape() as tape:
        tape.watch(channel_model1.parameter_list)
        choi_target = channel_target.choi(T)
        choi_model = channel_model1.choi(T)
        loss = tf.math.reduce_sum(tf.abs(choi_model - choi_target) ** 2)
    
    print(loss)
    loss1.append(loss.numpy())
    gradients = tape.gradient(loss, channel_model1.parameter_list)
    optimizer.apply_gradients(zip(gradients, channel_model1.parameter_list))
```

```python
n = 2
d = 2**n
T = 1

tf.random.set_seed(42)
np.random.seed(42)

H_target = SpinSpin()
H_model = SpinSpin()

channel_target = MagnusPropagator(liouvillian=H_target, grid_size=100)
channel_model2 = MagnusPropagator(liouvillian=H_model, grid_size=100)
print(state_fidelity(channel_target.choi(T), channel_model2.choi(T))/16)
```

```python
T = 0.1
loss2 = []
optimizer = tf.optimizers.Adam(learning_rate=0.05)

for j in tqdm(range(10)):
    for i in tqdm(range(100)):
        with tf.GradientTape() as tape:
            tape.watch(channel_model2.parameter_list)
            choi_target = channel_target.choi(T)
            choi_model = channel_model2.choi(T)
            loss = tf.math.reduce_sum(tf.abs(choi_model - choi_target) ** 2)

        print(loss)
        loss2.append(loss.numpy())
        gradients = tape.gradient(loss, channel_model2.parameter_list)
        optimizer.apply_gradients(zip(gradients, channel_model2.parameter_list))
    T += 0.1
```

```python
fig = plt.figure(figsize=(8,6), dpi = 300, facecolor='w', edgecolor='k')
plt.plot(loss1)
plt.plot(loss2)
plt.yscale("log")
plt.legend(["Constant T=1", "Increase T from 0.1 to 1"])
plt.xlabel("Iteration")
plt.ylabel("Loss")
plt.grid()
plt.show()
```

## Noise 

```python
n = 2
d = 2**n
T = 1

tf.random.set_seed(42)
np.random.seed(42)

H_model = SpinSpin(degree=3)
H_target = SpinSpin(degree=1)
jump_operator = JumpOperator(4)

lindblad_target_noisy = LindbladGenerator(hamiltonian = H_target, 
                                          jump_operator = jump_operator, 
                                          gamma = 0.025)

lindblad_target_ideal = LindbladGenerator(hamiltonian = H_target, 
                                          jump_operator = jump_operator, 
                                          gamma = 0.)

lindblad_model = LindbladGenerator(hamiltonian = H_model, 
                                   jump_operator = jump_operator, 
                                   gamma = 0.025)

channel_noisy = MagnusPropagator(liouvillian=lindblad_target_noisy, grid_size=100)
channel_ideal = MagnusPropagator(liouvillian=lindblad_target_ideal, grid_size=100)
channel_model = MagnusPropagator(liouvillian=lindblad_model, grid_size=100)
print(1- state_fidelity(channel_noisy.choi(T), channel_ideal.choi(T))/16)
```

```python
T = 0.1
#loss1 = []

for j in range(10):
    optimizer = tf.optimizers.Adam(learning_rate=0.01)
    for i in tqdm(range(200)):
        with tf.GradientTape() as tape:
            tape.watch(H_model.parameter_list)
            choi_target = channel_ideal.choi(T)
            choi_model = channel_noisy.choi(T)
            #loss = tf.math.reduce_sum(tf.abs(choi_model - choi_target) ** 2)
            loss = 1-state_fidelity(channel_model.choi(T), channel_ideal.choi(T))/16

        print(loss)
        loss1.append(loss.numpy())
        gradients = tape.gradient(loss, H_model.parameter_list)
        optimizer.apply_gradients(zip(gradients, H_model.parameter_list))
    T += 0.1
```
