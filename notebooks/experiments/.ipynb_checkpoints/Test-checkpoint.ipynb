```python
n = 3
d = 2**n

np.random.seed(42)
random.seed(42)
tf.random.set_seed(42)

kraus_target = KrausMap(d = d, 
                        rank = 4,
                        )

kraus_model = KrausMap(d = d, 
                       rank = d**2,
                      )

choi_target = maps_to_choi([kraus_target])
choi_model = maps_to_choi([kraus_model])

spectrum_target = choi_spectrum(choi_target, real=True)
spectrum_model = choi_spectrum(choi_model, real=True)
```

```python
print(spectrum_target.shape)
```

```python
model = ModelQuantumMap(
                        q_map = kraus_model,
                        loss = SpectrumDistance(sigma=0.1),
                        optimizer = tf.optimizers.Adam(learning_rate=0.01),
                        )
```

```python
model.train(inputs = [spectrum_target],
            targets = [],
            num_iter = 89,
            N = 1,
            use_batch = False,
            )
```

```python
choi_model = maps_to_choi([kraus_model])
spectrum_model = choi_spectrum(choi_model, real=True)

t = np.linspace(0, 2*np.pi, 1000)
circle = [np.sin(t), np.cos(t)]

fig=plt.figure(figsize=(8,8), dpi = 300, facecolor='w', edgecolor='k')
plt.plot(spectrum_target[:,0], spectrum_target[:,1], "o")
plt.plot(spectrum_model[:,0], spectrum_model[:,1], "o")
plt.plot(*circle)
plt.xlim([-1,1])
plt.ylim([-1,1])
plt.show()
```
