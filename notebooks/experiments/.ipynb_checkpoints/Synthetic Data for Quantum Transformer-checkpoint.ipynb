```python
import numpy as np
import matplotlib.pyplot as plt

from pickle import load, dump
```

# Brownian Motion

```python
def brownian_step(x, step_length):
    step = np.random.normal(0, step_length, size=(2,))
    x_new = x + step
    return x_new
```

```python
np.random.seed(42)

N = 100 #Number of time-series
T = 100 #Number of time steps
d = 2 # dimensionality of data
step_length = 0.1

x_series = np.zeros((N,T,d))
for n in range(N):
    for t in range(1, 100):
        x_series[n, t] = brownian_step(x_series[n, t-1], step_length)
    
plt.plot(*x_series[:10].T)
plt.show()
```

```python
dump(x_series, open("brownian_motion.p", "wb"))
```

# Logistic Map

```python
def logistic_map(x_seq, r = 1, k = 1):
    N = x_seq.shape[0]
    
    x_new = k*r*x_seq[-1]*(1-x_seq[-1])
    if N>1:
        x_new = x_new + (1-k)/(N-1)*np.sum(x_seq[:-1])
        
    return x_new
    
```

```python
N = 100 #Number of time-series
T = 300 #Number of time steps
d = 1 # dimensionality of data

x_series = np.zeros((N,T,d))

for n in range(N):
    x_series[n,:] = np.random.uniform(0,1,(T,1))
    for t in range(5,300):
        x_series[n,t] = logistic_map(x_series[n,t-5:t], r=5.5, k=0.5)
```

```python
plt.plot(x_series[0,250:])
plt.plot(x_series[1,250:])
plt.plot(x_series[2,250:])
plt.show()
```

```python
dump(x_series, open("logistic_map.p", "wb"))
```
